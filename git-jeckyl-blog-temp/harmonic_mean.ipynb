```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Harmonic mean

```python
R = np.arange(0.01,1,0.01) 
P = np.arange(0.01,1,0.01)
points = [[],[],[]]

for p in P:
    for r in R:
        h = 2*p*r/(p+r)

        points[0].append(p)
        points[1].append(r)
        points[2].append(h)
```

```python
fig, ax = plt.subplots(figsize=(4,3))
ax.scatter(points[0],points[1], c=points[2], cmap="gnuplot", s=5)
PCM=ax.get_children()[0]
plt.colorbar(PCM,label="F1-score", orientation="vertical", ax=ax)
plt.xlabel('Precision')
plt.ylabel('Recall')
fig.show()
```

## Arithmetic mean

```python
R = np.arange(0.01,1,0.01) 
P = np.arange(0.01,1,0.01)
points = [[],[],[]]

for p in P:
    for r in R:
        h = (p+r)/2

        points[0].append(p)
        points[1].append(r)
        points[2].append(h)
```

```python
fig, ax = plt.subplots(figsize=(4,3))
ax.scatter(points[0],points[1], c=points[2], cmap="gnuplot", s=5)
PCM=ax.get_children()[0]
plt.colorbar(PCM,label="F1-score", orientation="vertical", ax=ax)
plt.xlabel('Precision')
plt.ylabel('Recall')
fig.show()
```

## Geometric mean

```python
R = np.arange(0.01,1,0.01) 
P = np.arange(0.01,1,0.01)
points = [[],[],[]]

for p in P:
    for r in R:
        h = np.sqrt(p*r)

        points[0].append(p)
        points[1].append(r)
        points[2].append(h)
```

```python
fig, ax = plt.subplots(figsize=(4,3))
ax.scatter(points[0],points[1], c=points[2], cmap="gnuplot", s=5)
PCM=ax.get_children()[0]
plt.colorbar(PCM,label="F1-score", orientation="vertical", ax=ax)
plt.xlabel('Precision')
plt.ylabel('Recall')
fig.show()
```
